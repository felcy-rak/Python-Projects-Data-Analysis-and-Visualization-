```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
```

```python
#Load Datasets
apps_df=pd.read_csv(r'C:\Users\Felcy\Downloads\Play Store Data.csv')
reviews_df=pd.read_csv(r'C:\Users\Felcy\Downloads\User Reviews.csv')
```

```python
reviews_df
```

```python
#Merging Datasets on App
merge_df=pd.merge(reviews_df,apps_df[['App','Category','Rating']], on='App',how='inner')
```

```python
#Drop/Remove rows missing sentiment or rating values
merge_df=merge_df.dropna(subset=['Sentiment','Rating'])
```

```python
merge_df
```

```python
#Apps with more than 1000 reviews
app_review_counts=merge_df['App'].value_counts()
apps_more_1000=app_review_counts[app_review_counts > 1000].index
merge_filtered=merge_df[merge_df['App'].isin(apps_more_1000)]
```

```python
merge_filtered
```

```python
#Top 5 categories
top_categories= merge_filtered['Category'].value_counts().nlargest(5).index
merge_top= merge_filtered[merge_filtered['Category'].isin(top_categories)]
```

```python
merge_top
```

```python
#Rating Groups
def group_ratings(rating):
    if rating <=2:
        return '1-2 stars'
    elif rating <=4:
        return '3-4 stars'
    else:
        return '4-5 stars'


merge_top['Rating_Group'] = merge_top['Rating'].apply(group_ratings)
```

```python
merge_top
```

```python
#Group by category
grouped=merge_top.groupby(['Category','Rating_Group','Sentiment']).size().reset_index(name='Count')
```

```python
fig=px.bar(
    grouped,
    x='Category',
    y='Count',
    color='Sentiment',
    facet_col='Rating_Group',
    barmode='stack',
    color_discrete_map={'Positive': 'green','Neutral': 'gold','Negative': 'red'},
    labels={'Count':'Number of Reviews'},
    title='Sentiment Distribution by Rating Groups (Top 5 App Categories)',
    height=600
)

fig.update_layout(
    plot_bgcolor='black',
    paper_bgcolor='black',
    font_color='white',
    title_font={'size': 16},
    xaxis=dict(title_font={'size': 12}),
    yaxis=dict(title_font={'size': 12}),
    margin=dict(l=10, r=10, t=30, b=10)
)

def save_plot_as_html(fig,filename,description=""):
    fig.write_html(filename)
    print(f"Saved {filename}: {description}")
                   
                   
save_plot_as_html(fig,"Sentiment Distribution Graph.html","Sentiment Distribution across rating groups for top 5 categories.")
```

```python
import numpy as np
import plotly.graph_objects as go
from datetime import datetime
import pytz
```

```python
apps_df
```

```python
apps_df['Installs'] = apps_df['Installs'].astype(str).str.replace(r'[+,]', '', regex=True)
apps_df = apps_df[apps_df['Installs'].str.isnumeric()]
apps_df['Installs'] = apps_df['Installs'].astype(float)
apps_df['Price'] = apps_df['Price'].astype(str).str.replace('$', '', regex=False).astype(float)
```

```python
def size_to_mb(size):
    if pd.isna(size) or size == 'Varies with device':
        return np.nan
    if 'M' in size:
        return float(size.replace('M', ''))
    if 'k' in size:
        return float(size.replace('k', '')) / 1024
    return np.nan

apps_df['Size MB'] = apps_df['Size'].apply(size_to_mb)
```

```python
def android_ver(ver):
    try:
        return float(str(ver).split()[0])
    except:
        return np.nan

apps_df['Android_Ver_Num'] = apps_df['Android Ver'].apply(android_ver)
```

```python
filtered = apps_df[
    (apps_df['Installs'] >= 10000) &
    ((apps_df['Type'] == 'Free') | ((apps_df['Type'] == 'Paid') & (apps_df['Price'] * apps_df['Installs'] >= 10000))) &
    (apps_df['Android_Ver_Num'] > 4.0) &
    (apps_df['Size MB'] > 15) &
    (apps_df['Content Rating'] == 'Everyone') &
    (apps_df['App'].str.len() <= 30)
].copy()

filtered['Revenue'] = np.where(filtered['Type'] == 'Paid', filtered['Price'] * filtered['Installs'], 0)
```

```python
top_categories = (
    filtered.groupby('Category')['Installs']
    .mean()
    .sort_values(ascending=False)
    .head(3)
    .index
)

top_data = filtered[filtered['Category'].isin(top_categories)]
```

```python
grouped = (
    top_data
    .groupby(['Category', 'Type'])
    .agg({'Installs': 'mean', 'Revenue': 'mean'})
    .reset_index()
)
```

```python
def is_time_in_range(start_hour, end_hour, timezone='Asia/Kolkata'):
    now = datetime.now(pytz.timezone(timezone))
    return start_hour <= now.hour < end_hour
```

```python
if is_time_in_range(13, 14):

    fig = go.Figure()

    for app_type, color in zip(['Free', 'Paid'], ['skyblue', 'steelblue']):
        df_type = grouped[grouped['Type'] == app_type]
        fig.add_trace(go.Bar(
            x=df_type['Category'],
            y=df_type['Installs'],
            name=f'{app_type} - Installs',
            marker_color=color,
            yaxis='y1'
        ))
    
    for app_type, color in zip(['Free', 'Paid'], ['green', 'darkgreen']):
        df_type = grouped[grouped['Type'] == app_type]
        fig.add_trace(go.Scatter(
            x=df_type['Category'],
            y=df_type['Revenue'],
            name=f'{app_type} - Revenue',
            mode='lines+markers',
            marker=dict(color=color),
            yaxis='y2'
        ))

    fig.update_layout(
        title='Average Installs and Revenue for Free vs Paid Apps (Top 3 Categories)',
        plot_bgcolor='black',
        paper_bgcolor='black',
        font_color='white',
        title_font={'size': 16},
        margin=dict(l=10, r=10, t=30, b=10),
        yaxis=dict(title='Average Installs', titlefont=dict(size=12)),
        yaxis2=dict(title='Average Revenue ($)', overlaying='y', side='right', titlefont=dict(size=12)),
        xaxis=dict(title='Category', titlefont=dict(size=12)),
        legend=dict(x=0.01, y=0.99),
        height=600
    )

    save_plot_as_html(
        fig,
        "Installs_Revenue_Top3_Categories.html",
        "Installs vs Revenue for Free and Paid Apps (Top 3 Categories, Dual Axis)"
    )

else:
    print("The chart is only visible between 1 PM and 2 PM IST.")
```
