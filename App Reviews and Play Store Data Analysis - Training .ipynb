```python
def convert_size(size):
    if 'M' in size:
        return float(size.replace('M',''))
    elif 'k' in size:
        return float(size.replace('k',''))/1024
    else:
        return np.nan
app_df['Size']=app_df['Size'].apply(convert_size)
```

```python
app_df['Installs'].unique()
app_df = app_df[app_df['Installs'].str.contains(r'^\d+[,+]*$', na=False)]
app_df['Installs'] = app_df['Installs'].str.replace(',', '', regex=True).str.replace('+', '', regex=True).astype(int)
```

```python
import numpy as np
app_df['Log_Installs'] = np.log(app_df['Installs'])
```

```python

app_df = app_df[app_df['Reviews'].str.contains(r'^\d+[,+]*$', na=False)]
app_df['Reviews'] = app_df['Reviews'].str.replace(',', '', regex=True).str.replace('+', '', regex=True).astype(int)
import numpy as np
app_df['Log_Reviews'] = np.log(app_df['Reviews'])
```

```python
def rating_group(rating):
    if rating>=4:
        return 'Top rated app'
    elif rating>=3:
        return 'Above Average'
    elif rating>=2:
        return 'Average'
    else:
        return 'Below Average'
```

```python
app_df['Ratings_Group']=app_df['Rating'].apply(rating_group)
```

```python
import plotly.express as px
import plotly.io as pio
import webbrowser
import os
```

```python
html_files_path="./"
if not os.path.exists(html_files_path):
    os.makedirs(html_files_path)
```

```python
plot_contaniners=""
```

```python
def save_plot_as_html(fig,filename,insight):
    global plot_contaniners
    filepath=os.path.join(html_files_path,filename)
    html_content=pio.to_html(fig,full_html=False,include_plotlyjs='inline')
    plot_contaniners += f"""
    <div class = "plot-contaniner" id="{filename}" onclick ="openPlot('{filename}')">
        <div class="plot"> {html_content}</div> 
        <div class="insights">{insight}</div> 
    </div>
    """
    
    fig.write_html(filepath,full_html=False,include_plotlyjs='inline')
    
```

```python
plot_width=400
plot_height=300
plot_bg_color='black'
text_color='white'
title_font={'size':12}
axis_font={'size':12}
```

```python
# Figure 1
category_counts = app_df['Category'].value_counts().nlargest(10)

fig1 = px.bar(
    x=category_counts.index,
    y=category_counts.values,
    labels={'x': 'Category', 'y': 'Count'},
    title='Top Categories on Play Store',
    color=category_counts.index,
    color_discrete_sequence=px.colors.sequential.Plasma,
    width=400,
    height=300
)

fig1.update_layout(
    plot_bgcolor='black',
    paper_bgcolor='black',
    font_color='white',
    title_font={'size': 12},
    xaxis=dict(title_font={'size': 12}),
    yaxis=dict(title_font={'size': 12}),
    margin=dict(l=10, r=10, t=30, b=10)
)

save_plot_as_html(fig1, "Category Graph 1.html", "The top categories on the Play Store are dominated")



```
